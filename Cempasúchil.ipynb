```python
# import math
# normal leaf length to 1 on y axis
# center leaf width to 0 on x axis
# make a wider leaf

import math as math

s = 3.8 # y axis length in pixels

xf = (np.asarray(x_vals)/s)-0.386 # center x axis on 0
yf = (-np.asarray(y_vals)+3.8)/s # flip the leaf and set length to 1

xt = (np.asarray(xtip_vals)/s)-0.386 # center x axis on 0
yt = (-np.asarray(ytip_vals)+3.8)/s # flip the leaf and set length to 1


plt.fill(xf, yf, color="#FF9C00", alpha=1) # check that leaf is scaled correctly
plt.fill(xt, yt, color="#FF5B00", alpha=1) # check that leaf is scaled correctly
plt.axes().set_aspect('equal', 'box')
```

```python
# import golden angle
# set to radians

golden_angle = 137.5077640500378546463487396283702776206886952699253696312384958261062333851951
phi = golden_angle * ( math.pi / 180.0 ) 
```

```python
# calculate angles and radii of phyllotactic points beforehand

thetas = []
radii = []

for n in range(200):

    r = math.sqrt(n) 
    theta = n * phi 
    
    thetas.append(theta)
    radii.append(r)
```

```python
# import animation for python

from IPython.display import display, clear_output
import time  

# This loop is selecting a growing number of leaves
for i in range(200):

    xlist = []
    ylist = []
    
    # This selects the current set of thetas to use
    current_thetas = thetas[0:(i+1)]
    # This selects the current set of radii to use
    # Everytime the loop runs, this list will be recreated in the "forward" direction
    current_radii = radii[0:(i+1)]
    # After creating the list of radii, "revere" its order
    # Reversing the order of the radii means that the 
    # leaves first specified have the longest radius
    current_radii.reverse()

    # In this loop, for each of the current leaves, calculate x & y coordinates
    for n in range (i+1): 
        
        # Here, for each leaf calculate n caclulate r, theta, x, y, and append x & y to their lists
        # theta & r are calcualted by indexing using n the lists current_thetas and current_radii
        # x & y are calculated from theta & r, and then appended to xlist and ylist
        r = current_radii[n]
        theta = current_thetas[n]
                  
        x = r * math.cos(theta)
        y = r * math.sin(theta)
            
        xlist.append(x)
        ylist.append(y)
        
    # For animation, you need to call a figure
    fig = plt.figure(figsize=(10,10)) 
    fig = plt.figure(facecolor="black")
    
    for i in range(len(current_radii)):
        
        if current_radii[i] > 5: # create a maximum leaf length
            leaf_r = 5
        else:
            leaf_r = current_radii[i]
            
        if current_radii[i] > 5: # this is for old leaves to disappear using alpha
            alfa = 0
        elif current_radii[i] > 4.8:
            alfa = 0.1
        elif current_radii[i] > 4.7:
            alfa = 0.2
        elif current_radii[i] > 4.6:
            alfa = 0.3
        elif current_radii[i] > 4.5:
            alfa = 0.4
        elif current_radii[i] > 4.4:
            alfa = 0.5
        elif current_radii[i] > 4.3:
            alfa = 0.6
        elif current_radii[i] > 4.2:
            alfa = 0.7
        elif current_radii[i] > 4.1:
            alfa = 0.8
        elif current_radii[i] > 4.0:
            alfa = 0.9
        else:
            alfa=1

        leaf_theta = current_thetas[i]
        
        xffls = []
        yffls = []
        
        for j in range(len(xf)): # for each leaf, rotate and scale
            
            xff = leaf_r*(xf[j]*math.cos(leaf_theta) - yf[j]*math.sin(leaf_theta))
            yff = leaf_r*(yf[j]*math.cos(leaf_theta) + xf[j]*math.sin(leaf_theta))
            
            xffls.append(xff)
            yffls.append(yff)
            
        xttls = []
        yttls = []
        
        for k in range(len(xt)): # for each leaf, rotate and scale
            
            xtt = leaf_r*(xt[k]*math.cos(leaf_theta) - yt[k]*math.sin(leaf_theta))
            ytt = leaf_r*(yt[k]*math.cos(leaf_theta) + xt[k]*math.sin(leaf_theta))
            
            xttls.append(xtt)
            yttls.append(ytt)
        
        
        # plot the leaves!

        plt.fill(xffls, yffls, color="#FF5B00", alpha=alfa/1, zorder=i)
        plt.plot(xffls, yffls, color="#FFB700", alpha=alfa/1, zorder=i)
        plt.fill(xttls, yttls, color="#FF9C00", alpha=alfa/1, zorder=i)
        #plt.plot(xffls, yffls, color="#94B447", alpha=alfa, lw=1.3)
        plt.xlim(-5,5)
        plt.ylim(-5,5)
        
        plt.axes().set_aspect('equal', 'box')
        plt.axis('off')
        
        
    # This is the code that creates the animation
    plt.savefig("./cempaschil/"+ str(i)+".jpg",facecolor=fig.get_facecolor(), edgecolor='none')
    time.sleep(0.001) 
    clear_output(wait=True)
    display(fig)
    fig.clear()

# Closes the figure animation once complete
plt.close()

```
